```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = [20,30,40,50]
plt.plot(x,y)
plt.title("Example")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.axis([0,6,0,20])
plt.title("Example 2")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
x = [3,1,3]
y = [3,2,1]
plt.plot(x,y)
plt.title("Line Chart")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
a = [3,1,3,12,2,4,4]
b = [3,2,1,4,5,6,7]
plt.bar(a,b)
plt.title("Bar chart")
plt.legend(['items'])
plt.show()
```

```python
x = [1,2,3,4,5,6,7,3]
plt.hist(x, bins=[1,2,3,4,5,6,7])
plt.title("Histogram")
plt.show()
```

```python
a = [3,1,3,12,2,4,4]
b = [3,2,1,4,5,6,7]
plt.scatter(a,b)
plt.title("Scatter plot")
plt.show()
```

```python
import numpy as np
x = np.array([35,25,25,15])
cols = ["black", "#ebba36","#4bf280","#cd17d4"]
plt.pie(x, explode=(0,0.1,0,0), labels=["Maths", "English", "Science", "Social"], startangle=90, colors = cols)
plt.title("Pie Chart")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(z ,y ,x)
```

```python
import seaborn as sb
```

```python
data = sb.load_dataset("iris")
```

```python
data
```

```python
sb.lineplot(x="petal_length", y ="petal_width", data=data)
plt.title("Iris Dataset")
sb.set_style("dark")
plt.show()
```

```python
pal = sb.color_palette()
sb.palplot(pal)
plt.show()
```

```python
pal2 = sb.dark_palette("xkcd:golden", 8)
sb.palplot(pal2)
plt.show()
```

```python
sb.set_palette(pal2)
plt.subplot(211)
sb.lineplot(x="sepal_length", y="sepal_width", data=data)
sb.set_palette(pal)
plt.subplot(212)
sb.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

```python
total_sales = [3000, 2245, 1235, 5330, 4200]
explode = [0, 0, 0, 0.2, 0]

location = ['Bangalore', 'Delhi', 'Chennai', 'Mumbai', 'Kolkatta']
colors = sb.husl_palette(h=.5)
plt.pie(total_sales, labels=location, colors=colors,
        autopct='%.0f%%', explode=explode, shadow=True, rotatelabels='true')
plt.show()
```

```python
sb.barplot(x="species", y="petal_width", data=data)
sb.set_palette(sb.cubehelix_palette(hue=1))
plt.show()
```

```python
sb.histplot(x="species", y="sepal_width", data=data)
plt.show()
```

```python
sb.distplot(data["sepal_width"])
sb.set_palette(sb.husl_palette(l=.4))
plt.show()
```

```python
sb.jointplot(x="species", y="petal_length", data=data)
plt.show()
```

```python
sb.pairplot(data=data, hue="species")
plt.show()
```
